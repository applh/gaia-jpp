```python
def show_img (img):
    import cv2
    import IPython
    # correct color
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    _, png = cv2.imencode('.png', img)
    i = IPython.display.Image(data=png)
    IPython.display.display(i)
```

```python
# face detection in jpeg image
import cv2
import numpy as np
import os
import glob

curdir = os.getcwd()
print(curdir)

# get the file dir
cron_dir = os.path.abspath('../my-data/cron')
print(cron_dir)

# MATPLOTLIB
os.environ['MPLCONFIGDIR'] = cron_dir
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline


def face_detection (src):
    img = cv2.imread(src)
    # correct color channel
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # plt.imshow(img)
    # plt.show()

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)

    show_img(img)

    # plt.imshow(img, cmap='gray', interpolation='bicubic')
    # plt.show()

def face_eyes_detection (src):
    img = cv2.imread(src)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    # print('faces', faces)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        #print('roi_gray', roi_gray)
        #print('roi_color', roi_color)
        eyes = eye_cascade.detectMultiScale(roi_gray)
        #print('eyes', eyes)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)

    show_img(img)

src = '../my-data/cron/photo-1.jpg'
# pick a jpg file from ../my-data/cron/
list = glob.glob('../my-data/cron/photos/*.jpg')
# pick a random element in the list
src = np.random.choice(list)
print(src)
face_eyes_detection(src)
# face_detection(src)
```
